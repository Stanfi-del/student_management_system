## STUDENT RESULT MANAGEMENT SYSTEM

### Step 1: Create a Grade Function

```python
#Function to calculate grade

def calculate_grade(score):
    if 70 <= score <= 100:
        return "A"
    elif 60 <= score <= 69:
        return "B"
    elif 50 <= score <= 59:
        return "C"
    elif 45 <= score <= 49:
        return "D"
    else:
        return "F"
```

### Step 2: Create Dictionary to Store Records

```python
students = {}
```

### Step 3: Add Student Function

```python
def add_student():
    try:
        name = input("Enter student name: ")
        score = float(input("Enter score (0 - 100): "))
        if score < 0 or score > 100:
            print("Score must be between 0 and 100.")
            return
        students[name] = score
        print(f"{name} added successfully.")
    except ValueError:
        print("invalid score entered.")
```

### Step 4: View All Students

```python
def view_students():
    if not students:
        print("No student records found.")
        return
    print("\nSTUDENT RECORDS")
    print("-" * 40)
    for name, score in students.items():
        grade = calculate_grade(score)
        print(f"Name: {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")
        print("-" * 40)
```

### Step 5: Update Student Score

```python
def update_score():
    name = input("Enter student name: ")
    if name not in students:
        print("Student not found.")
        return
    try:
        new_score = float(input("Enter new score: "))
        if new_score < 0 or new_score > 100:
            print("Score must be between 0 and 100.")
            return
        students[name] = new_score
        print("Score updated successfully.")
    except ValueError:
        print("invalid score entered.")
            
```

### Step 6: Search Student

```python
def search_student():
    name = input("Enter student name to search: ")
    if name in students:
        score = students[name]
        grade = calculate_grade(score)
        print("\nStudent Found")
        print(f"Name : {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")
    else:
        print("Student not found.")
```

### Step 7: Class Statistics

```python
def class_statistics():
    if not students:
        print("No records available.")
        return
    scores = list(students.values())
    highest = max(scores)
    lowest = min(scores)
    average = sum(scores) / len(scores)
    print("\nCLASS STATISTICS")
    print("-" * 30)
    print("Highest Score:", highest)
    print("Lowest Score :", lowest)
    print("Average Score:", round(average,2))
    print("Total Students:", len(scores))
```

### Step 8: Main Menu Program

```python
while True:
    print("\n===== STUDENT RESULT MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Update Score")
    print("4. Search Student")
    print("5. Class Statistics")
    print("6. Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        update_score()
    elif choice == "4":
        search_student()
    elif choice == "5":
        class_statistics()
    elif choice == "6":
        print("program terminated.")
        break
    else:
        print("invalid option. Try again.")
```

## TASK 2B: INVENTORY MANAGEMENT SYSTEM

### Step 1: Create Inventory Dictionary

```python
inventory = {}
```

### Step 2: Add Product

```python
def add_product():
    product = input("Enter product name: ")
    try:
        quantity = int(input("Enter quantity: "))
        if quantity < 0:
            print("Quantity cannot be negative.")
            return
        inventory[product] = quantity
        print(f"{product} added successfully.")
    except ValueError:
        print("invalid quantity entered.")
```

### Step 3: View Inventory

```python
def view_inventory():
    if not inventory:
        print("Inventory is empty.")
        return
    print("\nINVENTORY RECORDS")
    print("-" * 40)
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
```

### Step 4: Update Product Quantity

```python
def update_quantity():
    product = input("Enter product name: ")
    if product not in inventory:
        print("Product not found.")
        return
    try:
        quantity = int(input("Enter new quantity: "))
        if quantity < 0:
            print("Quantity cannot be negative.")
            return
        inventory[product] = quantity
        print("Quantity updated successfully.")
    except ValueError:
        print("invalid quantity entered.")
```

### Step 5: Search Product

```python
def search_product():
    product = input("Enter product name: ")
    if product in inventory:
        print(f"{product}: {inventory[product]} units")
    else:
        print("Product not found.")
```

### Step 6: Remove Product

```python
def remove_product():
    product = input("Enter product name: ")
    if product in inventory:
        del inventory[product]
        print("Product removed successfully.")
    else:
        print("Product not found.")
```

### Step 7: Low Stock Report

```python
def low_stock_report():
    print("\nLOW STOCK PRODUCTS")
    found = False
    for product, quantity in inventory.items():
        if quantity < 5:
            print(product, "-", quantity)
            found = True
        if not found:
            print("No low-stock items.")
```

### Step 8: Inventory Statistics

```python
def inventory_statistics():
    if not inventory:
        print("Inventory is empty.")
        return
    total_products = len(inventory)
    highest_product = max(inventory, key=inventory.get)
    lowest_product = min(inventory, key=inventory.get)
    total_stock = sum(inventory.values())
    print("\nINVENTORY STATISTICS")
    print("-" * 35)
    print("Total Products:", total_products)
    print("Highest Quantity Product:", highest_product, "(", inventory[highest_product], ")"
         )
    print("Lowest Quantity Product:", lowest_product, "(", inventory[lowest_product], ")"
         )
    print("Total Stock Available:", total_stock)
```

### Step 9: Main Menu Program

```python
while True:
    print("\n===== INVENTORY MANAGEMENT SYSTEM =====")
    print("1. Add Product")
    print("2. View Inventory")
    print("3. Update Quantity")
    print("4. Search Product")
    print("5. Remove Product")
    print("6. Low Stock Report")
    print("7. Inventory Statistics")
    print("8. Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        add_product()
    elif choice == "2":
        view_inventory()
    elif choice == "3":
        update_quantity()
    elif choice == "4":
        search_product()
    elif choice == "5":
        remove_product()
    elif choice == "6":
        low_stock_report()
    elif choice == "7":
        inventory_statistics()
    elif choice == "8":
        print("Program terminated.")
        break
    else:
        print("Invalid option. Try again.")
```
